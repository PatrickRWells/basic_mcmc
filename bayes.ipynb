```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = [np.random.normal() for i in range(100)]
tot = 0
for index, val in enumerate(samples):
    tot += val*val

num = 100
```

```python
def p(a):
    fact = (np.log(a)/(2*np.pi))**(num/2)
    exp = a**(-tot/2)
    return fact*exp
    
```

```python
a = np.linspace(1,5,1000)
ys = p(a)
```

```python
plt.plot(a, ys)
```

```python
def draw_sample(xt):
    variance = 0.05
    return np.random.normal(xt, variance)
    

def walk_probability(xp, xt):
    variance = 0.05
    fact = 1/(np.sqrt(2*np.pi)*variance)
    exponent = -((xp - xt)**2)/(2*(variance**2))
    return fact*np.exp(exponent)

def check_acceptance(prob):
    
    if prob > 1:
        return True
    else:
        return True if prob >= np.random.uniform() else False
    


def mcmc(num = 100000):
    xt = 2.7
    xs = [xt]
    
    for i in range(num):
        xp = draw_sample(xt)
        if xp <= 1:
            xs = np.append(xs, xt)
            continue
            
        prob = p(xp)/p(xt)
        if check_acceptance(prob):
            xs = np.append(xs, xp)
            xt = xp
        else:
            xs = np.append(xs, xt)
    
    return xs
        
    
    
```

```python
chain = mcmc()
xs = [i+1 for i in range(len(chain))]
print(chain)
```

```python
plt.plot(xs, chain)
```
